```python
import matplotlib.pyplot as plt
import numpy as np
from random import random

#inverse function method
#n is number of samples
def inverseSampling(n):
    return [np.sin(random()/2*np.pi)**2 for _ in range(n)]

b=plt.hist(inverseSampling(100000), bins=100, density=True)
```

```python
#compare to real pdf
#hist should be area under this pdf, looks good
x=np.arange(0.00001,1,0.00001)
plt.plot(x,1/np.pi/np.sqrt(x*(1-x)))
```
