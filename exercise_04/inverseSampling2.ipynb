```python
import matplotlib.pyplot as plt
import numpy as np
import pylab
import scipy.stats as stats
from random import random
#1b
#inverse function method
#n is number of samples
def inverseSampling(n,a):
    r=[]
    for _ in range(n):
        u=random()
        if(u<=0.5):
            r.append(-(2*u)**(-a))
        else:
            r.append((-2*(u-1))**(-a))
    return r

b=plt.hist(inverseSampling(100000,0.4), bins=100, density=True)
```

```python
#1b
#compare to real pdf
#hist should be area under this pdf, looks good
x=np.arange(-50,-0.99,0.01)
a=0.4
plt.plot(x,(1/2/a)*abs(x)**(-1-1/a))

x=np.arange(1,60,0.01)
plt.plot(x,(1/2/a)*abs(x)**(-1-1/a))
```

```python
#1c
#n values of S_K
def partialSum(K,a,n):
    return[1/np.sqrt(K/(1-2*a))*np.sum(inverseSampling(K,a)) for _ in range(n)]

K=100
#probability plot
#N should be 10**6 but takes me tot long 
stats.probplot(partialSum(K,0.1,10**2), dist="norm", plot=pylab)
pylab.show()
stats.probplot(partialSum(K,0.25,10**2), dist="norm", plot=pylab)
pylab.show()
stats.probplot(partialSum(K,0.45,10**2), dist="norm", plot=pylab)
pylab.show()
```
