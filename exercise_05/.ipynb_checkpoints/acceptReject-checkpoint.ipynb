```python
import numpy as np
import matplotlib.pyplot as plt
from random import random
from scipy.optimize import curve_fit

#exercise 1, linear regression
#plot average annual temp. anomaly against years
#a)estimate errors (independent samples)

weather=np.loadtxt("globalTemp.csv",delimiter=",", skiprows=2)
years=[]
anomaly=[]
average=[]
squaredValue=[]
err=[]
plt.figure(figsize=(10,10))
#for 2019 not enough value
for i in range(len(weather)-1):
    years.append(weather[i,0])
    anomaly.append(np.average(weather[i,1:13]))
    average.append(weather[i,13])
    squaredValue=(np.average((weather[i,1:13])**2))
    err.append(np.sqrt(squaredValue-(weather[i,13])**2))
# or average instead of anomaly
#print(len(err))
#print(len(years))
#print(len(anomaly))
#print(err)
plt.errorbar(years,anomaly,yerr=err,ecolor='r')

#c) fitting polynom

def p3(f,a,b,c,d):
    return [a*x**3+b*x**2+c*x+d for x in f]

def p2(f,b,c,d):
    return [b*x**2+c*x+d for x in f]

def p1(f,c,d):
    return [c*x+d for x in f]

def p0(f,d):
    return [d for x in f]


popt, pcov = curve_fit(p3, years, anomaly)
plt.plot(years, p3(years, *popt), 'g--',label='degree 3: a=%5.3f, b=%5.3f, c=%5.3f, d=%5.3f' % tuple(popt))
popt, pcov = curve_fit(p2, years, anomaly)
plt.plot(years, p2(years, *popt), 'r--',label='degree 2: b=%5.3f, c=%5.3f, d=%5.3f' % tuple(popt))
popt, pcov = curve_fit(p1, years, anomaly)
plt.plot(years, p1(years, *popt), 'b--',label='degree 1: c=%5.3f, d=%5.3f' % tuple(popt))
popt, pcov = curve_fit(p0, years, anomaly)
plt.plot(years, p0(years, *popt), 'y--',label='const: d=%5.3f' % tuple(popt))
plt.xlabel('x')
plt.ylabel('y')
plt.legend(fontsize='x-large')
plt.show()

#degree 2 is enough
```

```python
#erxercise 2, acceprReject method
#true pdf
x=np.arange(-1,1.0001,0.0001)
f0=2*np.sqrt(1-x**2)/np.pi
plt.plot(x,f0)
plt.show()
#a) alpha=0

#N samples
def rejSample(N):
    samples=[]
    for _ in range(N):
        found=False 
        while not found:
            x=2*random()-1
            y=random()*2/np.pi
            if y<=2*np.sqrt(1-x**2)/np.pi:
                found=True
                samples.append(x)
    return samples
plt.hist(rejSample(100000), bins=100, density=True)
plt.title('task b')
plt.show()
#plt.set_title('task b')

#b) and d)
def rejSampleA(N,a):
    samples=[]
    wrong=0
    for _ in range(N):
        found=False 
        while not found:
            x=rejSample(1)[0]
            y=random()*np.sqrt(1-x**2)
            if y<=np.sqrt(1-x**2)*np.cos(a*x)**2:
                found=True
                samples.append(x)
            else:
                wrong=wrong+1
    return samples,N/(N+wrong)
# give explicit value for a, here a=3
Ars=[]
for a in range(1,17):
    samples,ar=rejSampleA(1000000,a)
    b=plt.hist(samples, bins=100, density=True)
    plt.show()
    Ars.append(ar)
    print(ar)
plt.plot(range(1,17),Ars)
plt.title("varying of AR with a")
```
