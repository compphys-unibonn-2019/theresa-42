```python
#task (a) and (b)
#autocorrelation

import numpy as np
import matplotlib.pyplot as plt
from random import random
from scipy.optimize import curve_fit

weather=np.loadtxt("globalTemp.csv",delimiter=",", skiprows=2)

quarters=[]
plt.figure(figsize=(10,10))
#for 2019 not enough value
for i in range(1,len(weather)-1):
    quarters.append(weather[i,15:])
    
#compute autocorrelation for the four quarts
#average for yeraly autocorrelation of anomaly
    
#convolution
def autocorr1(x):
    result = np.correlate(x,x, mode='full')
    return result[result.size/2:]
  
#statistical (normalised)
#could not handle negative t
#compute mean and variance only from overlapping time series
def autocorr2(x, t=1):
    t=abs(t)
    if t==0:
        return 1
    return np.corrcoef(np.array([x[:-t], x[t:]]))[1][0]
    
def autocorr3(x,t=1):
    t=abs(t)
    if t==0:
        return 1
    av=np.average(x)
    N=len(x)
    return np.sum((x[:-t]-av)*(x[t:]-av))/(N-t)/np.var(x)


quarters=np.array(quarters)
t=7
#auto1=np.average([autocorr1(quarters[:,i]) for i in range(4)])
auto2=np.average([autocorr2(quarters[:,i],t) for i in range(4)])
#plt.scatter()
auto3=np.average([autocorr3(quarters[:,i],t) for i in range(4)])

print(auto2,auto3)
x=range(20)
#x,acorr,n,m=plt.acorr(quarters[:,0],maxlags=5, label='acorr')
plt.plot(x,[np.average([autocorr2(quarters[:,i],t) for i in range(4)]) for t in x],label='auto2')
plt.errorbar(x,[np.average([autocorr3(quarters[:,i],t) for i in range(4)]) for t in x], yerr=1/len(quarters) ,label='auto3')
plt.xlabel('lag', fontsize=15)
plt.ylabel('correlation coefficient', fontsize=15)
plt.legend(fontsize='large')


#task (c)
#tau for t=0 is 1/2
tau=1/2
for t in range(1,len(quarters)):
    gamma=auto3=np.average([autocorr3(quarters[:,i],t) for i in range(4)])
    if gamma<=0:
        break
    tau+=gamma
print(tau)
```

```python
# task (c), regression towards the mean
#loc is mean, scale is sigma
#normally distribution

#(a)
#its a matrix, x is vector with components x1 and x2
x=np.random.normal(loc=0.0, scale= 1.0, size=(2,100000))
y=[x[0,i]*x[1,i] for i in range(100000)]
print(np.var(y))
print(np.mean(y))

#(b) prob. for y to be smaller than -1
#programming with constraints
sdY=np.sqrt(np.var(y))
print(len([a for a in y if a<-sdY])/100000)

#(c)
def c(ys,x1s,n):
    #how often y was smaller or greater than -sd(y)
    smaller=0
    greater=0
    sdY=np.sqrt(np.var(ys))
    sdx1=np.sqrt(np.var(x1s))
    for y,x1 in zip(ys,x1s):
        if x1<=n*sdx1:
            continue
        if y<-sdY:
            smaller+=1
        else:
            greater+=1
    if smaller+greater==0:
        return 'not defined'
    return smaller/(smaller+greater)
for n in range(5):
    print(str(n)+': '+ str(c(y,x[0,:],n)))
```
