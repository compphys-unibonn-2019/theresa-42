```python
import matplotlib.pyplot as plt
import numpy as np
from random import random

#example 2.5.1 of the script
def Z(n):
    return (np.sum([random() for _ in range (n)])-n/2.)*np.sqrt(12./n)

b=plt.hist([Z(12) for _ in range(10000)], bins=100, density=True)

#compute mean
# number of observations o

def mean(n,o):
    return np.average([Z(n) for _ in range(o)])
#print(12,10000)

#compute variance
def variance(n,o):
    return np.average([Z(n)**2 for _ in range(o)])-mean(n,o)**2

```

```python
plt.plot(range(1,26),[mean(n,10000) for n in range (1,26)])
plt.plot(range(1,26),[variance(n,10000) for n in range (1,26)])
```
